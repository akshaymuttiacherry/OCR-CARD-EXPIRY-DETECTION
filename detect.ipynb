```python
import pytesseract
import cv2
```

```python
img=cv2.imread('sbi.jpg')
img
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
```

```python
pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
text=pytesseract.image_to_string(img)
text
```

```python
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
print(data)
```

```python
import re
```

```python
for t in data ['text']:
    if re.search(r'\b(0[1-9]|1[0-2])[\/\-](\d{2}|\d{4})\b',t):
        print(t)
```

```python
n = len(data['text'])

for i in range(n):

    text = data['text'][i].strip()
    conf = float(data['conf'][i])

    if conf > 30 and re.search(r'\b(0[1-9]|1[0-2])/\d{2}\b', text):

        x = int(data['left'][i])
        y = int(data['top'][i])
        w = int(data['width'][i])
        h = int(data['height'][i])

        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()


```

```python
from datetime import datetime

expiry_date = None
for t in data['text']:
    if re.search(r'\b(0[1-9]|1[0-2])/\d{2}\b', t):
        expiry_date = t.strip()
        break


if expiry_date:

    month, year = expiry_date.split('/')
    month = int(month)
    year = int("20" + year)   # convert 30 → 2030

    now = datetime.now()

    if year > now.year or (year == now.year and month >= now.month):
        print("✅ Card is VALID")
    else:
        print("❌ Card is EXPIRED")

else:
    print("Expiry date not detected")

```
